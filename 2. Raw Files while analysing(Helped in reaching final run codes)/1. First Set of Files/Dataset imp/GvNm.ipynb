```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
G = nx.Graph()

nodes = []
edges = np.loadtxt('football2.txt')
for i in edges:
    for j in i:
        if j not in nodes:
            nodes.append(int(j))
G.add_nodes_from(nodes)
G.add_edges_from(edges)

nx.draw(G, with_labels = True , node_color = 'r')
```

```python
NUM_ITERATIONS = 20


for i in range(NUM_ITERATIONS):
    edge_betweenness = nx.edge_betweenness_centrality(G).items()
    edge_to_delete = sorted(edge_betweenness, key = lambda pair: -pair[1])[0][0]
    
    G.remove_edge(*edge_to_delete)
    
    nx.draw(G, with_labels = True , node_color = 'r')
    plt.title('Step %s\nEdge %s Deleted'%(i, edge_to_delete))
    
    plt.show()
```
