```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms.community.centrality import girvan_newman
import networkx.algorithms.community as nx_comm

G = nx.karate_club_graph()
communities = girvan_newman(G)
node_groups = []
for com in next(communities):
    node_groups.append(list(com))

print(node_groups)

color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append("red")
    else:
        color_map.append("orange")

modularity = nx_comm.modularity(G, node_groups)
print("Mod: ", modularity)
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from networkx.algorithms.community.centrality import girvan_newman
import networkx.algorithms.community as nx_comm

nodes = []
edges = np.loadtxt('football.txt')
for i in edges:
    for j in i:
        if j not in nodes:
            nodes.append(int(j))
            
graph=nx.Graph() 
graph.add_nodes_from(nodes) #adds nodes
graph.add_edges_from(edges) #add edges

G = graph
# G = nx.karate_club_graph()
modularity = []
# for i in range(0,10):
communities = girvan_newman(G)
node_groups = []
for com in next(communities):
    node_groups.append(list(com))
print(node_groups)
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append("red")
    else:
        color_map.append("orange")

modularity.append(nx_comm.modularity(G, node_groups))
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
print("Mod: ", min(modularity))

```
